```python
# ==========================================================
# AI MULTI-MODEL ENSEMBLE PREDICTION SYSTEM
# ==========================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.datasets import make_classification

np.random.seed(42)

# ----------------------------------------------------------
# 1. GENERATE COMPLEX CLASSIFICATION DATA
# ----------------------------------------------------------

X, y = make_classification(
    n_samples=3000,
    n_features=15,
    n_informative=10,
    n_redundant=3,
    n_classes=2,
    random_state=42
)

df = pd.DataFrame(X)
df["Target"] = y

print("Dataset Shape:", df.shape)
print(df.head())

# ----------------------------------------------------------
# 2. TRAIN TEST SPLIT
# ----------------------------------------------------------

X = df.drop("Target", axis=1)
y = df["Target"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2
)

# ----------------------------------------------------------
# 3. INDIVIDUAL MODELS
# ----------------------------------------------------------

lr = LogisticRegression(max_iter=1000)
rf = RandomForestClassifier(n_estimators=150)
gb = GradientBoostingClassifier()

lr.fit(X_train, y_train)
rf.fit(X_train, y_train)
gb.fit(X_train, y_train)

# ----------------------------------------------------------
# 4. ENSEMBLE MODEL (VOTING)
# ----------------------------------------------------------

ensemble = VotingClassifier(
    estimators=[
        ('lr', lr),
        ('rf', rf),
        ('gb', gb)
    ],
    voting='hard'
)

ensemble.fit(X_train, y_train)

# ----------------------------------------------------------
# 5. MODEL COMPARISON
# ----------------------------------------------------------

models = {
    "Logistic Regression": lr,
    "Random Forest": rf,
    "Gradient Boosting": gb,
    "Ensemble Model": ensemble
}

results = {}

for name, model in models.items():
    preds = model.predict(X_test)
    acc = accuracy_score(y_test, preds)
    results[name] = acc

print("\nModel Accuracy Comparison:")
for k, v in results.items():
    print(k, ":", round(v, 4))

# ----------------------------------------------------------
# 6. CONFUSION MATRIX (ENSEMBLE)
# ----------------------------------------------------------

ensemble_preds = ensemble.predict(X_test)

cm = confusion_matrix(y_test, ensemble_preds)

plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt='d', cmap="Blues")
plt.title("Ensemble Confusion Matrix")
plt.show()

# ----------------------------------------------------------
# 7. ACCURACY VISUALIZATION
# ----------------------------------------------------------

plt.figure(figsize=(8,5))
plt.bar(results.keys(), results.values())
plt.xticks(rotation=45)
plt.title("Model Accuracy Comparison")
plt.ylabel("Accuracy")
plt.show()

# ----------------------------------------------------------
# 8. LIVE PREDICTION INTERFACE
# ----------------------------------------------------------

print("\n------ LIVE INTELLIGENT DECISION SYSTEM ------")
print("Enter 15 feature values:")

user_input = []
for i in range(15):
    value = float(input(f"Feature {i+1}: "))
    user_input.append(value)

user_array = np.array(user_input).reshape(1, -1)

prediction = ensemble.predict(user_array)[0]
prob = np.mean([
    lr.predict_proba(user_array)[0][1],
    rf.predict_proba(user_array)[0][1],
    gb.predict_proba(user_array)[0][1]
])

if prediction == 1:
    print("\n⚠ High Risk / Positive Class Detected")
else:
    print("\n✅ Low Risk / Negative Class")

print("Ensemble Confidence Score:", round(prob * 100, 2), "%")
```
